```python
letters = ['a','b','c','d','e','f','g','h','i','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']

choice = input('''Do you want to Encrypt or Decrypt?
Type (e/d): ''').lower()
key = input("Enter keyword: ").lower().replace('j', 'i')
text = input("Enter the text: ").lower().replace('j', 'i')

v = []
pt1 = []
for i in key:
    if i not in v:
        v.append(i)
        pt1.append(i)

for i in letters:
    if i not in v:
        v.append(i)
        pt1.append(i)

pt2 = [pt1[i:i+5] for i in range(0, 25, 5)]

def find_pos(ch):
    for i in range(5):
        for j in range(5):
            if pt2[i][j] == ch:
                return i, j
    return -1, -1

def prepare_digraphs(txt, mode):
    digraphs = []
    i = 0
    while i < len(txt):
        a = txt[i]
        b = ''
        if i+1 < len(txt):
            b = txt[i+1]
            if a == b:
                b = 'x'
                i += 1
            else:
                i += 2
        else:
            b = 'x'
            i += 1
        digraphs.append(a + b)
    if mode == 'd':
        digraphs = [txt[i:i+2] for i in range(0, len(txt), 2)]
    return digraphs

def playfair_cipher(pairs, mode):
    result = []
    for pair in pairs:
        a, b = pair[0], pair[1]
        r1, c1 = find_pos(a)
        r2, c2 = find_pos(b)

        if r1 == r2:
            if mode == 'e':
                result.append(pt2[r1][(c1+1)%5] + pt2[r2][(c2+1)%5])
            else:
                result.append(pt2[r1][(c1-1)%5] + pt2[r2][(c2-1)%5])
        elif c1 == c2:
            if mode == 'e':
                result.append(pt2[(r1+1)%5][c1] + pt2[(r2+1)%5][c2])
            else:
                result.append(pt2[(r1-1)%5][c1] + pt2[(r2-1)%5][c2])
        else:
            result.append(pt2[r1][c2] + pt2[r2][c1])
    return result

if choice == 'e':
    digraphs = prepare_digraphs(text, 'e')
    result = playfair_cipher(digraphs, 'e')
    print("\nEncrypted Ciphertext:", ''.join(result))
elif choice == 'd':
    digraphs = prepare_digraphs(text, 'd')
    result = playfair_cipher(digraphs, 'd')
    print("\nDecrypted Plaintext:", ''.join(result))
else:
    print("Invalid choice!")
```
